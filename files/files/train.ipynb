# WebGPU Trainer
Train a GPT entirely in your browser.

Run each cell in order.

```javascript
return (async () => {
  const mod = await import("/files/trainer/engine.js");
  globalThis.createEngine = mod.createEngine;
  return "Engine loaded";
})();
```

```javascript
return (async () => {
  globalThis.losses = [];
  globalThis.engine = await createEngine({
    onStep: (loss, step) => { globalThis.losses.push(loss); }
  });
  return "Engine ready";
})();
```

```javascript
return (async () => {
  const STEPS = 100;
  for (let i = 0; i < STEPS; i++) {
    await engine.step(i);
    if (i % 10 === 0 || i === STEPS-1) {
      process.stdout.write("Step " + String(i).padStart(3) + ": loss=" + losses[losses.length-1].toFixed(4) + "
");
    }
  }
  const early = losses.slice(0,10).reduce((a,b)=>a+b)/10;
  const late  = losses.slice(-10).reduce((a,b)=>a+b)/10;
  return "Loss: " + early.toFixed(4) + " -> " + late.toFixed(4) + " (" + ((1-late/early)*100).toFixed(1) + "% improvement)";
})();
```

```javascript
return (async () => {
  const text = await engine.sample("Once upon a time", 200);
  return text;
})();
```
